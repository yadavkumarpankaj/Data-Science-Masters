## Imputation technique

## 1 Mean value inputation

```python
sns.distplot(df.age)
```

```python
df.age.isnull().sum()
```

```python
df['Age_mean'] = df['age'].fillna(round(np.mean(df.age),2))
```

```python
df
```

```python
df[['age','Age_mean']]
```

- this technique work well when your data is normally distributed

## 2 - Median Value Imutation

```python
df['age_median'] = df['age'].fillna(round(df.age.median(),2))
```

```python
df[['age','age_median','Age_mean']]
```

## 3- Mode Value Imputation --> categorical features

```python
df.embarked.isnull()
```

```python
df[df.embarked.isnull()]
```

```python
df.embarked.unique()
```

```python
mode = df[df.age.notna()].embarked.mode()[0]
```

```python
mode
```

```python
df['embarked_mode']  = df.embarked.fillna(mode)
```

```python
df.embarked_mode.isnull().sum()
```
